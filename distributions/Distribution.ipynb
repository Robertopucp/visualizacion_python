```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
enaho = pd.read_stata("data/data_enaho.dta")
enaho
```

```python
enaho['l_salario'].plot(kind = 'hist', bins = 10, figsize = (8,6))
plt.title('Logaritmo del salario por hora')
plt.show()
```

```python
enaho['l_salario'].plot(kind = 'hist', bins = 30, figsize = (8,6))
plt.title('Logaritmo del salario por hora')
plt.show()
```

```python
sns.set('paper')
sns.kdeplot(enaho['l_salario'], label = 'Densidad')
plt.title('Logaritmo del salario por hora')
plt.show()
```

```python

sns.kdeplot(data = enaho, x = "l_salario", bw_adjust=  0.2)
plt.title('Logaritmo del salario por hora')
plt.show()
```

```python
sns.kdeplot(enaho['l_salario'], hue = "empresa", fill=True, common_norm=False, palette="crest", alpha=0.5, linewidth=0.2)
plt.title('Logaritmo del salario por hora')
plt.show()
```

```python
sns.kdeplot(data = enaho, x = "l_salario", hue = "tipo_hogar" )
plt.title('Logaritmo del salario por hora')
plt.show()
```
